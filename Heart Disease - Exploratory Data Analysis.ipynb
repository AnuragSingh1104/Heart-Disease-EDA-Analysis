```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head(10)
```

## Distribution plot for Numeric Column

```python
sns.distplot(df['Age'], kde=True, hist = True,rug='True')
```

```python
sns.distplot(df['Age'], bins = 30)
```

```python
sns.displot(df['RestingBP'], kde=True,rug=True,color ='blue')
```

```python
sns.distplot(df['Cholesterol'], kde=True,color ='blue')
```

```python
sns.distplot(df['MaxHR'], kde=True,color ='blue')
```

## Pie chart for Categorical

```python
df.groupby('Sex').size().plot(kind='pie',autopct= '%.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct= '%.1f')
```

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct= '%.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind='pie',autopct= '%.1f')
```
